#### freq
: freq 인수로 특정한 날짜만 생성되도록 할 수도 있음
- s: 초 / T: 분 / H: 시간
- D: 일(day)
- B: 주말이 아닌 평일
- W: 주(일요일) / W-MON: 주(월요일)
- M: 각 달(month)의 마지막 날 / MS: 각 달의 첫날
- BM: 주말이 아닌 평일 중에서 각 달의 마지막 날 / BMS: 주말이 아닌 평일 중에서 각 달의 첫날
- WOM-2THU: 각 달의 두번째 목요일
- Q-JAN: 각 분기의 첫달의 마지막 날 / Q-DEC: 각 분기의 마지막 달의 마지막 날

```python
# 2023-1-1 ~ 2023-2-16 사이 평일만 추출
pd.date_range("2023-1-1", "2023-2-16", freq="B")
```

```python
# 2023-1-1 ~ 2023-2-16 사이 월요일만 추출
pd.date_range("2023-1-1", "2023-2-16", freq="W-MON")
```

```python
# 2022-1-1 ~ 2023-2-16 사이 각 분기 첫달의 마지막 날 추출
pd.date_range("2022-1-1", "2023-2-16", freq="Q-JAN")
```

### First/Last
- 시계열 데이터에서 처음/끝 값으로부터 특정 기간만큼의 데이터를 필터링하는 메서드
- df.first/last(offset) : offset은 기준이 될 기간

```python
# 2017년 데이터에서 5일까지만 추출
crime_2017 = crime_df.loc['2017'].sort_index()
crime_2017.first('5D') # 기간 = 5일
```

```python
# 2017년 데이터 마지막에서 3일까지만 추출
crime_2017.last('3D') # 기간 = 3일
```

### Resample
- Datetime Index를 원하는 주기로 나누어주는 메서드
- 시간 간격을 재조정
  - 시간 간격 작아지면 데이터 양 증가해서 up-sampling
  - 시간 간격 커지만 데이터 양 감소해서 down-sampling

```python
crime_2017
```

```python
# 2017년 데이터에서 주별 평균 추출
crime_2017.resample('W').mean()
```

```python
# 2017년 데이터에서 월별 데이터 추출
crime_2017.resample('M').first() # first로 (첫번째)값 보여줌
```

```python
# 전체 데이터에서 연도별 최소값 추출
crime_df.resample('Y').min()
```

### KNN

```python
# 국가가 UK인 데이터에 대한 새 컬럼 추가
result = [] 
for i in  sales_new['Country']: 
    if i == 'United Kingdom':
      result.append(1) 
    else:
      result.append(0)        
sales_new['UK'] = result

sales_new.head()
```

```python
# X, y 나누기
X = sales_new[['UnitPrice', 'Quantity']]
y = sales_new['UK']
```

```python
# train/test 데이터셋 나누기
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = y, random_state = 42)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
# KNN 알고리즘 적용 : 대표적인 머신러닝 모델
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 3) # 참조할 이웃의 수
knn.fit(X_train, y_train)
```

```python
# 예측 및 스코어 도출
knn.predict(X_test)
pred_test = knn.predict(X_test)

from sklearn.metrics import f1_score
f1 = f1_score(y_test, pred_test)
print("f1 score:\n", f1)
```

```python
# 분류예측 레포트
from sklearn.metrics import classification_report
cfreport = classification_report(y_test, pred_test)
print("분류예측 레포트:\n", cfreport)
```
