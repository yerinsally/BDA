## melt
- 지정한 열의 데이터를 모두 행으로 정리
- 사용하는 특정한 경우가 있고, 그때 데이터 전처리를 수월하게 할 수 있음
- 엑셀의 피벗과 비슷) 열 데이터를 행으로 들어가게 만들거나 깔끔하지 못한 컬럼을 정리하여 깔끔한 데이터로 만든다



- id_vars : 위치를 그대로 유지하는 열의 이름
- value_vars : 행으로 위치를 변경할 열의 이름
- var_name : value_var 위치로 변경한 열의 이름
- value_name : var_name 위치로 변경한 열의 데이터 컬럼 이름

```python
ebola = pd.read_csv('/content/drive/MyDrive/[BDAA 5기] 데이터 분석 기초반/데이터셋/country_timeseries.csv')
ebola
```

```python
ebola_new = pd.melt(ebola, id_vars=['Date','Day'])
ebola_new
```

```python
variable = ebola_new.variable.str.split('_')
variable
```

```python
# 새로운 컬럼 추가하기기
status_va =  variable.str.get(0) # 첫번째 열
country_va =  variable.str.get(1) # 두번째 열열

ebola_new['country'] = country_va
ebola_new['status'] = status_va
ebola_new
```
