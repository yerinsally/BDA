# 데이터 전처리 

- 데이터 전처리 data preprocessing 가장 많은 시간을 사용

- 다양한 데이터를 가지고 분석하기 전에 전처리가 필요함
- 분석기초반은 다양한 데이터를 최대한 다뤄보며 데이터 전처리 역량을 키울 수 있음

- pandas 데이터 분석 전처리 작업에서 가장 많이 사용되는 패키지!

데이터 전처리 관련 함수

- groupby() : 집단별 나누기
- agg() : 통계치 구하기
- merge(): 데이터 합치기 (열)
- concat(): 데이터 합치기 (행)

## 4. assign()

- assign() 이용해서 파생변수를 만들기
- 그냥 바로 직관적으로 새로운 컬럼을 만들 수 있다.

```python
# exam 데이터프레임 다시 불러오기
exam = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/BDAA 5기/data/exam.csv')
```

```python
exam.assign(tot = exam['math'] + exam['english'] + exam['science'])
```

```python
exam['total'] = (exam['math'] + exam['english'] + exam['science'])
exam['mean'] = (exam['math'] + exam['english'] + exam['science']) / 3
exam
```

```python
exam_new = exam.sort_values('total',ascending = False)
exam_new
```

람다식을 이용해서 데이터 전처리

- lambda x:

```python
# exam 데이터프레임 다시 불러오기
exam = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/BDAA 5기/data/exam.csv')
```

```python
exam.assign(tot = lambda x : x['math']+ x['english']+x['science'])
```

```python
exam.assign(tot = lambda x: x['math'] + x['english'] + x['science'],
           mean = lambda x: x['tot']/3)
```

## 5. groupby()

- .groupby('컬럼') : 해당 컬럼을 묶어서 보여줌
- .agg(변수 = ('컬럼', '원하는 통계치')

```python
# 반별로 특정 컬럼의 통계치(평균) 구하기
exam.groupby('nclass').agg(mean_math = ('math','mean'))
```

```python
exam.groupby('nclass').agg(sum_math = ('math','sum'))
```

```python
# 여러 개의 변수의 요약 통계
exam.groupby('nclass').agg(mean_math = ('math','mean'),
                          sum_math = ('math','sum'),
                          median_math = ('math','median'),
                          n = ('nclass','count'))
```

```python
# groupby() : 빠르게 요약 통계 구할 수 있음
exam.groupby('nclass').sum()
```

## 6. merge(), concat()
- 데이터 병합 및 합치기

```python
a = pd.DataFrame({'id':[1,2,3,4,5],
                 'midterm':[50,60,70,80,90]})

b = pd.DataFrame({'id':[1,2,3,4,5],
                 'final':[40,60,80,30,80]})
```

```python
display(a)
display(b)
```

```python
total = pd.merge(a, b, how = 'inner')
total
```

```python
# concat : axis(0 = 행, 1 = 열)
pd.concat([a,b], axis = 0) # 행을 기준으로 합치기
```

```python
pd.concat([a,b],axis = 1) # 열을 기준으로 합치기
```
