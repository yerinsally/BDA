### 데이터 불러오기
1. 고객 기본 정보
- age: 나이 (numeric)
- job: 직업분류 (“admin.”,“unknown”,“unemployed”,“management”,“housemaid”,“entrepreneur”,“student”, “blue-collar”,“self-employed”,“retired”,“technician”,“services")
- marital: 결혼상태 (“married”,“divorced”,“single”) - “divorced” means divorced or widowed
- education: 교육정도 (“primary”,“secondary”,“tertiary”, “unknown”)
- default: 채무불이행(연체)여부 (“yes”,“no”)
- balance: 연평균잔고, in euros (numeric)
- housing: 주택대출여부 (“yes”,“no”)
- loan: 개인대출여부 (“yes”,“no”)
2. 현 캠패인에서 마지막 접촉 관련
- contact: 접촉 채널 (“unknown”,“telephone”,“cellular”)
- day: 마지막 접촉날짜(일) (numeric)
- month: 마지막 접촉날짜(월) (“jan”, “feb”, “mar”, …, “nov”, “dec”)
- duration: 마지막 접촉에서 지속 시간(초) (numeric)
3. 기타 속성:
- campaign: 본 캠페인 기간 고객 접촉 건수 (numeric, includes last contact)
- pdays: 직전 캠페인에서 접촉 후 경과 일수 (numeric, -1 means client was not - - previously contacted)
- previous: 본 캠페인 이전 접촉 건수 (numeric)
- poutcome: 직전 캠페인에서의 결과 (“unknown”,“other”,“failure”,“success”)
4. Output variable (desired target):
- y: 정기예금(term deposit) 가입 여부 (“yes”,“no”)

### concat

```python
pd.concat([bank, bank_add], ignore_index=True)
```

```python
pd.concat([bank, bank_add], ignore_index=True, axis=0)
```

```python
pd.concat([bank, bank_add], axis=1)
```

### merge

```python
bank['age'].value_counts()
```

```python
pd.merge(bank, bank_add, left_on=['age'], right_on=['age'])
```

```python
pd.merge(bank, bank_add, left_index=True, right_on='age')
```

```python
pd.merge(bank, bank_add, on='age', how='inner') # 교집합
```

```python
pd.merge(bank, bank_add, on='education')
```

```python
pd.merge(bank, bank_add, on='age', how='right')
```

```python
# 다양한 키로 병합합
pd.merge(bank, bank_add, on=['age','job','marital','education','default','housing','loan'], how='right')
```

```python
pd.merge(bank, bank_add, on=['age','job','marital','education','default','housing','loan'], how='left')
```
