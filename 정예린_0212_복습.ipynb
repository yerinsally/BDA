## datetime
- datetime.date(연, 월, 일)
- datetime.time(시간, 분, 초, 마이크로초)
- datetime.datetime : 날짜와 시간
- datetime.timedelta : 일수와 시간
- Timestamp : 날짜와 시간을 모두 가지고 있는 단일 객체


```python
import pandas as pd
import datetime
```

```python
date = datetime.date(year=2023, month=2, day=12)
time = datetime.time(hour=23, minute=32, second=5, microsecond=1234)
```

```python
print(date, time)
```

```python
# datetime.datetime
dt = datetime.datetime(year=2023, month=2, day=12,hour=23, minute=32, second=5, microsecond=1234)
print(dt)
```

```python
# timedelta
td = datetime.timedelta(days=5, hours=5)
print(td)
```

```python
# timedelta로 시계열적 계산 가능능
print(date + td) # 2023년 2월 12일 + 5일
print(dt + td) # 시간까지 더해줌
```

## to_datetime 
- pandas에서 제공하는 함수
- 시계열데이터로 변환해주는 방법
- Date 형식의 자료형을 datetime 오브젝트로 변환

```python
pd.to_datetime('2023-2-12')
# 연도/월 잘못입력하면 알아서 에러 찾아줌
```

```python
a = pd.Series([10,20,30,40])
pd.to_datetime(a, unit='D') # 1970 1월 1일
```

### 시계열 데이터를 인덱스로 변경!

```python
# .set_index('시계열 데이터')
crime_df = crime.set_index('REPORTED_DATE')
crime_df
```

```python
# 시계열 데이터를 인덱스로 변경 후
# 슬라이싱, 원하는 요일, 연도 등 자유롭게 추출 가능
crime_df.loc['2017']
```

```python
crime_df.loc['2017-09-12']
```

```python
crime_df.loc['Dec 2015']
```

```python
crime_df.loc['2017'].sort_index()
```

```python
crime_df.loc['2017-01-01':'2017-01-12'].sort_index()
```

```python
# between_time : 지정한 시간 사이 데이터 추출
crime_df.between_time('10:00','15:00').sort_index()
```

```python
# at_time : 원하는 시간만 추출
crime_df.at_time('17:12').sort_index()
```

필수과제 : 다양한 문법 찾기

```python
# first
crime_df_sr = crime_df.at_time('17:12').sort_index()
crime_df_sr.first('5D')
```

```python
# resample : 월별, 일별, 주별 값을 계산
crime_df.resample('Y').sum()
```

```python
crime_df.resample('Y').max()
```
