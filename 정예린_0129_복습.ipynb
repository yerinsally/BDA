# 데이터 병합

## concat 

```python
import pandas as pd
df1 = pd.DataFrame({'A':[1,2,3],
                    'B':[4,5,6]})
df2 = pd.DataFrame({'A':[7,8,9], 
                    'B':[10,11,12]})
display(df1)
display(df2)
```

```python
# axis = 0 : 행 방향
pd.concat([df1,df2], axis=0)
```

```python
# axis = 1 : 열 방향
pd.concat([df1,df2], axis=1) 
```

```python
# 인덱스를 재정렬 할 경우
pd.concat([df1,df2], axis=0, ignore_index=True)
```

```python
# 차원이 다른 경우
df1 = pd.DataFrame({'A':[1,2,3],
                   'B':[4,5,6]})
df2 = pd.DataFrame({'A':[7,8],
                   'B':[10,11]})
display(df1)
display(df2)
```

```python
df1.shape
df2.shape
```

```python
pd.concat([df1,df2], axis=1)
```

## merge 함수

- left : 병합할 첫 번째 왼쪽 데이터 프레임
- right: 병합할 두 번쨰 오른쪽 데이터 프레임
- on : 조인 키로 사용할 컬럼 목록, 입력된 컬럼은 꼭 두 데이터프레임에 모두 공통으로 포함 되어야함
- left_on : 왼쪽 데이터프레임에서 조인 키로 사용할 컬럼 목록
- right_on : 오른쪽 데이터프레임에서 조인 키로 사용할 컬럼 목록
- left_index : 왼쪽 데이터프레임에서 조인 기준으로 인덱스를 사용할지
- right_index: 오른쪽 데이터프레임에서 조인 기준으로 인덱스 사용할지
- how: 조인 방법을 결정 (디폴트는 inner)

```python
df1 = pd.DataFrame({'A':[1,2,3,4],
                    'B':[5,6,7,8]})
df2 = pd.DataFrame({'A':[1,2,3,4],
                    'C':[10,11,12,13]})
display(df1)
display(df2)
```

```python
# on 인자 사용하기
pd.merge(df1, df2, on='A')
```

```python
pd.merge(df1, df2, on='B') # Error
```

조인키를 사용할 컬럼을 한 개 이상으로 하게 되는 경우?
- 고객데이터 예시 : 고객 ID or 주민등록번호 (고유함, 중복이 없음)
- 만약 주문 테이블에 고객 ID가 없다면, 두 개의 컬럼 사용해야함.
- 주문날짜와 주문제품번호 조인키로 사용

```python
df1 = pd.DataFrame({'A':[1,2,3,4],
                    'B':[5,6,7,8],
                    'D':[0,0,0,3]})
df2 = pd.DataFrame({'A':[1,2,3,4],
                    'B':[5,6,7,8],
                    'E':[1,1,1,0]})
display(df1)
display(df2)
```

```python
# 조인키 두개
pd.merge(df1, df2, on=['A','B'])
```

### left_on , right_on

```python
df1 = pd.DataFrame({'A':[1,2,3,4],
                    'B':[5,6,7,8],
                    'D':[0,0,0,3]})
df2 = pd.DataFrame({'a':[1,2,3,4],
                    'B':[5,6,7,8],
                    'E':[1,1,1,0]})
display(df1)
display(df2)
```

```python
pd.merge(df1, df2, left_on=['A'], right_on=['a'])
```

```python
pd.merge(df1, df2, left_on=['A','B'], right_on=['a','B'])
```

### left_index, right_index

```python
df1 = pd.DataFrame({'A':[1,2,3,4],
                    'D':[0,0,0,3]})
df2 = pd.DataFrame({'A':[1,2,3,4],
                    'E':[1,1,1,0]})
display(df1)
display(df2)
```

```python
# 인덱스 설정할 수 있음
# 인덱스를 설정하는 방법은 컬럼.set_index('컬럼', inplace=True)
# 인덱스를 리셋하는 방법은 컬럼.reset_index('컬럼', inplace=True)
df1.set_index('A',inplace=True)
```

```python
pd.merge(df1, df2, left_index=True, right_on='A')
```

```python
pd.merge(df1, df2, left_index=True, right_on='E')
```

### How
- inner : 교집합 (default)
- right : 오른쪽 데이터프레임 조인 키 병합
- left : 왼쪽 데이터프레임 조인 키 병합
- outer : 합집합 

```python
df1 = pd.DataFrame({'A':[1,2,3,4],
                    'D':[0,0,0,3]})
df2 = pd.DataFrame({'A':[1,2,7,4],
                    'E':[1,1,1,0]})
display(df1)
display(df2)
```

```python
pd.merge(df1, df2, on='A', how='inner') # 교집합
```

```python
pd.merge(df1, df2, on='A')
```

```python
pd.merge(df1, df2, on='A', how='left')
```

```python
pd.merge(df1, df2, on='A', how='right')
```

```python
pd.merge(df1, df2, on='A', how='outer')
```
