## 행 단위 데이터 추출
- loc : 인덱스를 기준으로 행 데이터 추출, loc[행 인덱싱 값, 열 인덱싱 값]
- iloc  : 행 번호를 기준으로 헹 데이터 추출, loc[행 인덱스, 열 인덱스]



```python
# gapminder : 국가별 경제 수준과 의료 수준 동향을 정리한 DataSet
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/BDAA 5기/data/gapminder.tsv',sep = '\t')
df
```

# 간단한 시각화
- 시각화는 본인의 분석 목적에 맞게 그래프를 그려야 한다.

```python
# 시각화 리이브러리 불러오기
import matplotlib.pyplot as plt
```
