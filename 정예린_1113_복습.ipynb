```python
# mpg 데이터에 cty 변수 활용한 test 변수 만들기
mpg['cty'].plot.hist() # 빈도 시각화
```

```python
import numpy as np
mpg['test'] = np.where(mpg['cty'] >= 20, 'good', 'bad')
mpg.head()
```

```python
mpg['test'].value_counts()
```

```python
# test bad / good 빈도
test_count = mpg['test'].value_counts()
test_count
```

```python
test_count.plot.bar() # 시각화
```
